```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#데이터 생성
np.random.seed(42)
X = np.random.rand(100, 1)
y = 4 + 3 * X + np.random.rand(100, 1)

w = np.random.randn(1)
b = np.random.randn(1)
y_study = w * X + b

#학습 전 그래프
#생성된 데이터 plotting
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 1, 0, 15])

plt.plot(X, y_study, "r-")
plt.show()
```

```python
learning_rate = 0.01 #학습률
n_try = 10000 #반복횟수
n = len(X) #데이터 개수

# Gradient Descent Algorithm
for Try in range(n_try):
    y_study = w * X + b
    error = y_study - y
    w_Differ = (2/n) * np.sum(error * X)
    b_Differ = (2/n) * np.sum(error)

    w -= learning_rate * w_Differ
    b -= learning_rate * b_Differ

# 학습 후 그래프
#생성된 데이터 plotting
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 1, 0, 15])

plt.plot(X, y_study, "r-")
plt.show()
```
